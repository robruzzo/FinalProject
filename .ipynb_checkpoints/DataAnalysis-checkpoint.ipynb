```python
"""
Robert E Ruzzo III
DataAnalysis.ipynb

The purpose of this notebook is to get an idea what the composition of the data is

"""
import pandas as pd
import os
import matplotlib.pyplot as plt
import random
from PIL import Image
```

```python
#Load the Labels and peform a count
data = pd.read_csv('D:\\Datasets\\histopathologic-cancer-detection\\train_labels.csv')
train_path = 'D:\\Datasets\\histopathologic-cancer-detection\\train'
data['label'].value_counts()
```

```python
#Calculate the compostion of the data
neg = 130908/220025
pos = 89117/220025
print("The dataset is composed of " + str(neg) +"% Negative and "+str(pos)+"% positive images.")
```

```python
#Create a chart showing some examples of each category
fig, ax = plt.subplots(2,3, figsize=(8,5))
fig.suptitle('Example Histopathological Images',fontsize=15)
# No Tumor Present
for i, idx in enumerate(data[data['label'] == 0]['id'][:3]):
    path = os.path.join(train_path, idx)
    ax[0,i].imshow(Image.open(path + '.tif'))
ax[0,0].set_ylabel('No Tumor', size='large')
# Tumor Present
for i, idx in enumerate(data[data['label'] == 1]['id'][:3]):
    path = os.path.join(train_path, idx)
    ax[1,i].imshow(Image.open(path + '.tif'))
ax[1,0].set_ylabel('Tumor Images', size='large')
```
